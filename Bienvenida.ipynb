## Formato de curso 

Este curso está dirigido a todos aquellos que no poseen conocimientos ni experiencia programando en Python o en cualquier otro lenguaje. Está orientado a la visualización y manejo de datos aunque se aprenderá, lo fundamental de este lenguaje para tomar el camino que cada uno eliga. 

Luego de este curso podrán utilizar la programación en tareas diarias para volverlas más eficaces, ahorrar tiempo, y  disminuir errores. 

Es un curso de tipo MOOC (Massive Open Online Course) y su diseño está basado en metolodogías ya probadas y estudiadas para incentivar y mejorar el auto-aprendizaje. [Metodologia AMMIL](https://www.researchgate.net/publication/338105863_AMMIL_A_methodology_for_developing_video-based_learning_courses). Las plataformas y la distribución de contenidos también se basan en el curso de Python dictado por nuestros compañeros de la Rama Estudiantil del Instituto Técnico de Buenos Aires [Página Rama](https://www.instagram.com/ieee.itba/?hl=es-la) y cursos open source:
